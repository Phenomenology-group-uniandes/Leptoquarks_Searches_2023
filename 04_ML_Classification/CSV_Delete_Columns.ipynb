```python
columns_to_delete = {"hadronic_dLQ": hadronic_dLQ,
        "hadronic_sLQ": hadronic_sLQ, 
        "hadronic_non-resonant": hadronic_non_resonant, 
        "semileptonic_dLQ": semileptonic_dLQ, 
        "semileptonic_sLQ": semileptonic_sLQ, 
        "semileptonic_non-resonant": semileptonic_non_resonant}
```

```python
bkgs = ['ttbar', 'stop', 'z_jets', 'w_jets',  'ww', 'wz', 'zz', 'Z_jets_LQ_DATA']

Masses = ['1250', '1500', '1750', '2000', '2250', '2500']
signals = ['LQ_LQ', 'Tau_LQ', 'Tau_Tau']
```

```python
Data = {}
```

```python
for signal in signals:
    for M in Masses:
        for channel in channels:
            ruta = os.path.join(os.sep,"disco4","pheno_csv_files","Leptoquarks_Searches", f"{signal}_{M}", f"{signal}_{M}_{channel}.csv")
            Data[ruta] = pd.read_csv(ruta)
            
for bkg in bkgs:
    for channel in channels:
        ruta = os.path.join(os.sep,"disco4","pheno_csv_files","Leptoquarks_Searches", f"{bkg}", f"{bkg}_{channel}.csv")
        Data[ruta] = pd.read_csv(ruta)
```

```python
for key in Data.keys():
    for channel in channels:
        if channel in key: 
            for column in columns_to_delete[channel]: Data[key] = Data[key].loc[:, Data[key].columns != column]
```
