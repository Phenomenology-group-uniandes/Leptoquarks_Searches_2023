```python
import os
import pickle

path = os.path.join(os.getcwd(), "data", "significances_curves")
```

```python
import numpy as np


def Calcular_g_U(c_U, M):
    v = 246
    return 2 * np.sqrt(c_U) * M * 1000 / v
```

```python
colors = {'137': 'red', '300': 'blue', '3000': 'green'}
keys = colors.keys()
case = "woRHC"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

if case == "wRHC":
    c_u1, c_u2 = 0.006, 0.002
else:
    c_u1, c_u2 = 0.017, 0.007
x_min = 0.5  # TeV

if case == "wRHC":
    x_max = 5.0  # TeV
else:
    x_max = 5.0  # TeV

x_prime = np.linspace(x_min, x_max, 100)
x_1, y_1 = x_min, Calcular_g_U(c_u1, x_min)
x_2, y_2 = x_max, Calcular_g_U(c_u1, x_max)

x_3, y_3 = x_min, Calcular_g_U(c_u2, x_min)
x_4, y_4 = x_max, Calcular_g_U(c_u2, x_max)

y1 = np.poly1d(np.polyfit([x_1, x_2], [y_1, y_2], deg=1))(x_prime)
y2 = np.poly1d(np.polyfit([x_3, x_4], [y_3, y_4], deg=1))(x_prime)

x = np.concatenate((x_prime[0], x_prime, x_prime[-1]), axis=None)
y = np.concatenate((y2[0], y1, y2[-1]), axis=None)

N = 25

sigma_curves = {}
for key in keys:
    with open(os.path.join(path, case, "combined", f"curves_{key}fb.pkl"), "rb") as f:
        sigma_curves[key] = pickle.load(f)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x_prime, y1, c="black", alpha=0.5)
ax.plot(x_prime, y2, c="black", alpha=0.5)
ax.fill(x, y, "gray", alpha=0.3, label=r"$1 \sigma R_{D^{(*)}}$")
ax.set_xlim(x_min, x_max)
ax.set_ylim(0.5, 3.5)

betard_title = {
    "wRHC": r"$, \mathbf{\beta_{R}}$ = -1",
    "woRHC": r"$, \mathbf{\beta_{R}}$ = 0",
}

for n, key in enumerate(keys):
    curves = sigma_curves[key]
    X1 = np.concatenate(
        [points.vertices[:, 0] for points in curves.collections[0].get_paths()]
    )
    Y1 = np.concatenate(
        [points.vertices[:, 1] for points in curves.collections[0].get_paths()]
    )
    data1 = pd.DataFrame(Y1, X1)

    if key == "dLQ":
        data1 = pd.DataFrame([0.25, 3.5], [min(data1.index), max(data1.index)])
    else:
        data1 = data1.sort_index(ascending=True)
        data1 = data1[~data1.index.duplicated(keep="first")]
        CS = CubicSpline(data1.index, data1.values)
        data1 = pd.DataFrame(
            CS(np.linspace(min(data1.index), max(data1.index), N)),
            np.linspace(min(data1.index), max(data1.index), N),
        )

    X2 = np.concatenate(
        [points.vertices[:, 0] for points in curves.collections[1].get_paths()]
    )
    Y2 = np.concatenate(
        [points.vertices[:, 1] for points in curves.collections[1].get_paths()]
    )
    data2 = pd.DataFrame(Y2, X2)
    if key == "dLQ":
        data2 = pd.DataFrame([0.25, 3.5], [min(data2.index), max(data2.index)])
    else:
        data2 = data2.sort_index(ascending=True)
        data2 = data2[~data2.index.duplicated(keep="first")]
        CS = CubicSpline(data2.index, data2.values)
        data2 = pd.DataFrame(
            CS(np.linspace(min(data2.index), max(data2.index), N)),
            np.linspace(min(data2.index), max(data2.index), N),
        )

    X3 = np.concatenate(
        [points.vertices[:, 0] for points in curves.collections[2].get_paths()]
    )
    Y3 = np.concatenate(
        [points.vertices[:, 1] for points in curves.collections[2].get_paths()]
    )
    data3 = pd.DataFrame(Y3, X3)
    if key == "dLQ":
        data3 = pd.DataFrame([0.25, 3.5], [min(data3.index), max(data3.index)])
    else:
        data3 = data3.sort_index(ascending=True)
        data3 = data3[~data3.index.duplicated(keep="first")]
        CS = CubicSpline(data3.index, data3.values)
        data3 = pd.DataFrame(
            CS(np.linspace(min(data3.index), max(data3.index), N)),
            np.linspace(min(data3.index), max(data3.index), N),
        )

    if n == 0:
        ax.legend(loc="upper left", frameon=False)
        plt.plot(0, 0, label=r"1.69 $\sigma$", linestyle="dashed", c="black")
        plt.plot(0, 0, label=r"3 $\sigma$", linestyle="dashdot", c="black")
        plt.plot(0, 0, label=r"5 $\sigma$", linestyle="dotted", c="black")
    plt.plot(0, 0, color=colors[key], label=f"${key}fb^{{-1}}$")
    plt.plot(data1, linestyle="dashed", color=colors[key])
    plt.plot(data2, linestyle="dashdot", color=colors[key])
    plt.plot(data3, linestyle="dotted", color=colors[key])

plt.legend(fontsize=11, loc="lower right",  ncol=2)
plt.xlim(1, x_max)
plt.ylim(0.5, 3.5)

plt.title(
    r'$\mathbf{\sqrt{s}}$ = 13.6 TeV'
    + betard_title[case],
    loc="right",
    fontsize=12,
    fontweight="bold",
)
plt.xlabel(r"$M_{U}$ [TeV]", fontsize=20)
plt.ylabel(r"$g_U$", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig(
    os.path.join(path, case, f"significances_13.6Tev.pdf"), bbox_inches="tight"
)
```
